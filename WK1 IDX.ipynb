```python
import pandas as pd

# File names
listing_file = "CRMLSListing202401-202603.csv"
sold_file = "CRMLSSold202401-202603.csv"

def process_residential(file_path, output_name, label):
    print(f"--- Processing {label} ---")
    
    # Reading raw data
    # Confirm row counts before filtering
    df = pd.read_csv(file_path, low_memory=False)
    
    initial_rows = len(df)
    print(f"Total rows before filter: {initial_rows}")

    # PropertyType = Residential

    if 'PropertyType' in df.columns:
        filtered_df = df[df['PropertyType'] == 'Residential'].copy()
        print(f"Rows after Residential filter: {len(filtered_df)}")
        
        filtered_df.to_csv(output_name, index=False)
        print(f"Saved to {output_name}\n")
        return filtered_df
    else:
        print(f"Error: Column 'PropertyType' not found in {label}!\n")
        return None

# Run and capture output
df_listings = process_residential(listing_file, "Combined_Listings_Residential.csv", "Listings")
df_sold = process_residential(sold_file, "Combined_Sold_Residential.csv", "Sold")
```
